```python
import os
import numpy as np
import pandas as pd
```

```python
path = 'D:/Projects/keras-ocr/datasets/NewDatasets/by_class'
folderToFind = "by_class/"
offsetLength = len(folderToFind)
trainFiles = []
validateFiles = []
```

```python
# r=root, d=directories, f = files
for r, d, f in os.walk('.'):
    for file in [fx for fx in f if fx.endswith(".png")]:
        filePath = os.path.join(r, file)
        fileClass = filePath[filePath.find(folderToFind) + offsetLength]
        if filePath.find('\\train\\') > -1:
            trainFiles.append({
                "fileName": filePath,
                "class": fileClass,
            })
        elif filePath.find('\\validate\\') > -1:
            validateFiles.append({
                "fileName": filePath,
                "class": fileClass,
            })
```

```python
for dirpath, dirnames, filenames in os.walk("."):
    for filename in [f for f in filenames if f.endswith(".png")]:
        print (os.path.join(dirpath, filename))
```

```python
print("Training Files: %d, Validation Files: %d" % (len(trainFiles), len(validateFiles)))
```

```python
dft = pd.DataFrame(trainFiles)
dft.to_csv("trainRecords.csv", encoding='utf-8', index=False)

dfv = pd.DataFrame(validateFiles)
dfv.to_csv("validateRecords.csv", encoding='utf-8', index=False)
```
